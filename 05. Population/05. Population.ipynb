## 1. 인구 데이터 확보하고 정리하기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import platform
from matplotlib import font_manager, rc

plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
else:
    print('Unknown System..')
```

```python
population = pd.read_excel('../data/05. population_raw_data.xlsx', header=1)  # 두번째 줄부터 읽기
population.fillna(method='pad', inplace=True)  # 빈 셀에 대해 NaN 처리를 하지 않고, 그 앞 내용으로 채우기

# 적절히 컬럼의 이름 바꾸기
population.rename(columns={'행정구역(동읍면)별(1)':'광역시도',
                            '행정구역(동읍면)별(2)':'시도',
                            '계':'인구수'}, inplace=True)

population = population[(population['시도'] != '소계')]  # '소계'항목 삭제

population
```

```python
population.is_copy = False

population.rename(columns= {'항목':'구분'}, inplace=True)

# 긴 이름 바꾸기
# '총인구수 (명)', '구분'] = '합계' => 여기에서 '구분'을 빼고 입력하면 전체 숫자항목까지 다 '합계'로 바뀌어버림
# '구분'을 넣어줘야 '구분'부분만 바꾼다는 걸 명시할 수 있음
population.loc[population['구분'] == '총인구수 (명)', '구분'] = '합계'
population.loc[population['구분'] == '남자인구수 (명)', '구분'] = '남자'
population.loc[population['구분'] == '여자인구수 (명)', '구분'] = '여자'

population
```

## 2. 인구 소멸 위기 지역 계산하고 데이터 정리하기

```python
# 인구 소멸 위기 지역을 알기 위해서는 20-30대의 인구와 65세 이상 인구수도 알아야 함
population.head()
```

```python
population['20-39세'] = population['20 - 24세'] + population['25 - 29세'] + \
                        population['30 - 34세'] + population['35 - 39세']
population['65세이상'] = population['65 - 69세'] + population['70 - 74세'] + \
                        population['75 - 79세'] + population['80 - 84세'] + \
                        population['85 - 89세'] + population['90 - 94세'] + \
                        population['95 - 99세'] + population['100+']
                    
population.head(10)
```

```python
# 광역시도와 시도를 index로 잡고, 인구수와 20-39세, 65세 이상만 데이터로 가져오기
pop = pd.pivot_table(population, index=['광역시도', '시도'], columns=['구분'], values=['인구수', '20-39세', '65세이상'])
pop
```

```python
# 인구 소멸 비율 계산하기
pop['소멸비율'] = pop['20-39세', '여자'] / (pop['65세이상', '합계'] / 2)
pop.head()
```

```python
# 소멸비율이 1이하이면 소멸위기지역
pop['소멸위기지역'] = pop['소멸비율'] < 1.0
pop.head()
```

```python
pop[pop['소멸위기지역']==True].index.get_level_values(1)
```

```python
# pivot_table에 의해 다단으로 구성된 index를 초기화
pop.reset_index(inplace=True)
pop.head()
```

```python
# 다단으로 표시된 column을 하나로 합치기
tmp_columns = [pop.columns.get_level_values(0)[n] + pop.columns.get_level_values(1)[n] 
                for n in range(0,len(pop.columns.get_level_values(0)))]
pop.columns = tmp_columns
pop.head()
```

```python
pop.info()
```

## 3. 지도 시각화를 위해 지역별 고유 ID 만들기

```python
pop['시도'].unique()
```

```python
si_name = [None] * len(pop)

# 광역시가 아니면서 구를 가지고 있는 시와 행정구를 dic형으로 선언
tmp_gu_dict = {'수원' : ['장안구', '권선구', '팔달구', '영통구'],
                '성남' : ['수정구', '중원구', '분당구'],
                '안양' : ['만안구', '동안구'],
                '안산' : ['상록구', '단원구'],
                '고양' : ['덕양구', '일산동구' ,'일산서구'],
                '용인' : ['처인구', '기흥구', '수지구'],
                '청주' : ['상당구', '서원구', '흥덕구', '청원구'],
                '천안' : ['동남구', '서북구'], 
                '전주' : ['완산구', '덕진구'],
                '포항' : ['남구', '북구'],
                '창원' : ['의창구', '성산구', '진해구' ,'마산합포구', '마산회원구'],
                '부천' : ['오정구', '원미구', '소사구']}
```

```python
for n in pop.index:

    # 광역시도의 있는 이름의 끝 세 글자가 '광역시', '특별시', '자치시'로 끝나지 않으면 일반 시, 군으로 분류
    if pop['광역시도'][n][-3:] not in ['광역시', '특별시', '자치시']:

        # 강원도와 경상남도에 동일한 이름을 가진 '고성군'이 있어서 처리
        if pop['시도'][n][:-1] == '고성' and pop['광역시도'][n] == '강원도':
            si_name[n] = '고성(강원)'
        elif pop['시도'][n][:-1] == '고성' and pop['광역시도'][n] == '경상남도':
            si_name[n] = '고성(경남)'

        # 일반 시에서 구를 가지는 경우에 대해 대응    
        else:
            si_name[n] = pop['시도'][n][:-1]

        # 일반 시에서 구를 가지는 경우에 대해 대응
        for keys, values in tmp_gu_dict.items():
            if pop['시도'][n] in values:
                if len(pop['시도'][n]) == 2:
                    si_name[n] = keys + ' ' + pop['시도'][n]
                elif pop['시도'][n] in ['마산합포구', '마산회원구']:
                    si_name[n] = keys + ' ' + pop['시도'][n][2:-1]
                else:
                    si_name[n] = keys + ' ' + pop['시도'][n][:-1]

    # 세종특별자치시도 세종으로 처리
    elif pop['광역시도'][n] == '세종특별자치시':
        si_name[n] = '세종'

    # 나머지 광역시도에서 앞 두 글자만 선택
    else:
        if len(pop['시도'][n]) == 2:
            si_name[n] = pop['광역시도'][n][:2] + ' ' + pop['시도'][n]
        else:
            si_name[n] = pop['광역시도'][n][:2] + ' ' + pop['시도'][n][:-1]
```

```python
si_name
```

```python
# 결과를 pop에 포함
pop['ID'] = si_name
```

```python
# 큰 의미없는 컬럼 제거 
del pop['20-39세남자']
del pop['65세이상남자']
del pop['65세이상여자']

pop.head()
```
