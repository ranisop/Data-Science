## 2-2. 좀 더 편리한 시각화 도구 - Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style('whitegrid')

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
# Seaborn에서 제공되는 기본적인 데이터셋
# Tips 데이터셋 : 요일별 점심, 저녁, 흡연여부, 식사금액, 팁을 정리한 데이터
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
# hue 옵션 사용
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='Set3')
plt.show()
```

```python
# lmplot : scatter처럼 그리고 직선으로 regression한 그림도 같이 그려주고, 유효범위도 ci로 잡아준다.
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips, size=7)
plt.show()
```

```python
# lmplot도 hue 옵션을 가질 수 있으며 미리 준비된 palette로 색상을 지정할 수 있다.
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips, palette='Set1', size=7)
plt.show()
```

```python
# flights 데이터셋 : 연도 및 월별 항고기 승객수 기록한 데이터
flights = sns.load_dataset('flights')
flights.head()
```

```python
# pivot으로 간편하게 월별, 연도별로 구분
flights = flights.pivot('month', 'year', 'passengers')
flights.head()
```

```python
# heatmap
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt='d')
plt.show()
```

```python
# iris 데이터셋 : 아이리스 꽃에 대한 데이터(꽃잎, 꽃받침의 너비와 폭을 가지고 그 종을 구분)
sns.set(style='ticks')
iris = sns.load_dataset('iris')
iris.head()
```

```python
# pairplot
sns.pairplot(iris, hue='species')
plt.show()
```
